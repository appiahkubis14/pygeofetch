````python
import pygeovision as pgv
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
import warnings; warnings.filterwarnings('ignore')

BASE_DIR = Path("./outputs/15_air_quality"); BASE_DIR.mkdir(parents=True,exist_ok=True)
BBOX = (12.35, 41.80, 12.65, 42.00)
client = pgv.PyGeoVision()
print(client); print("\nStudy: Rome, Italy | Sentinel-5P TROPOMI")
print("WHO limits: NO2 < 40 ug/m3  |  PM2.5 < 5 ug/m3")
````

````python
s5p = client.search(bbox=BBOX,date_range=("2024-01-01","2024-12-31"),
    providers=["planetary_computer"],collections=["sentinel-5p-l2-no2"],
    cloud_cover_max=100,limit=20)
print(f"Sentinel-5P granules: {len(s5p)}")
for r in s5p[:3]: print(f"  {r.date}")

MONTHS  = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
NO2     = [85,78,62,45,38,28,22,25,41,68,82,88]
PM25    = [18,16,13,10, 8, 7, 6, 7, 9,14,17,19]
ann_no2 = sum(NO2)/12; ann_pm25 = sum(PM25)/12

print(f"\nMonthly NO2 (umol/m2)  Annual mean: {ann_no2:.0f}")
print(f"Monthly PM2.5 (ug/m3)  Annual mean: {ann_pm25:.1f} (WHO limit: 5)")
print(f"Months NO2>70 (alert)  : {sum(1 for v in NO2 if v>70)}/12")
print(f"Months PM2.5>15 (unhealthy): {sum(1 for v in PM25 if v>15)}/12")
````

## Visualisation

````python
fig,axes = plt.subplots(1,2,figsize=(13,5))
c1=['#E74C3C' if v>70 else ('#F39C12' if v>45 else '#27AE60') for v in NO2]
axes[0].bar(MONTHS,NO2,color=c1,edgecolor='white')
axes[0].axhline(70,color='red',ls='--',lw=2,label='Alert (70)')
axes[0].set_ylabel("NO2 (umol/m2)"); axes[0].set_title("Monthly NO2 — Rome 2024",fontsize=11,fontweight='bold')
axes[0].legend(); axes[0].set_ylim(0,100)
c2=['#E74C3C' if v>15 else ('#F39C12' if v>10 else '#27AE60') for v in PM25]
axes[1].bar(MONTHS,PM25,color=c2,edgecolor='white')
axes[1].axhline(5,color='black',ls=':',lw=1.5,label='WHO (5 ug/m3)')
axes[1].axhline(15,color='red',ls='--',lw=2,label='Unhealthy (15)')
axes[1].set_ylabel("PM2.5 (ug/m3)"); axes[1].set_title("Monthly PM2.5 — Rome 2024",fontsize=11,fontweight='bold')
axes[1].legend()
plt.suptitle("Air Quality — Rome, Italy (Sentinel-5P TROPOMI)",fontsize=12,fontweight='bold')
plt.tight_layout(); plt.savefig(BASE_DIR/"air_quality.png",dpi=150,bbox_inches='tight'); plt.show()
````
