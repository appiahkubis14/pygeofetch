## 4. Search Commands — All 16 Flags

```python
!PyGeoFetch search run --help
```

```python
# Basic table output
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --start-date 2024-01-01 \
    --end-date 2024-03-01 \
    --cloud-cover 0-20 \
    --providers aws_earth \
    --format table
```

```python
# All filters — satellite, sort, max results
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --start-date 2024-01-01 \
    --end-date 2024-06-01 \
    --cloud-cover 0-10 \
    --providers planetary_computer \
    --satellites Sentinel-2 \
    --sort-by cloud_cover \
    --sort-order asc \
    --max-results 5 \
    --format table
```

```python
# Multi-provider, save to GeoJSON
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --start-date 2024-01-01 \
    --end-date 2024-06-01 \
    --cloud-cover 0-15 \
    --providers aws_earth,planetary_computer,element84 \
    --max-results 20 \
    --output output/cli_results.geojson \
    --format table
print()
!ls -lh output/cli_results.geojson
```

```python
# JSON output — machine-readable
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --start-date 2024-01-01 --end-date 2024-03-01 \
    --cloud-cover 0-10 \
    --providers aws_earth \
    --max-results 3 \
    --format json
```

```python
# CSV output
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --cloud-cover 0-10 \
    --providers aws_earth \
    --max-results 5 \
    --format csv
```

```python
# IDs only — for scripting and pipelines
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --cloud-cover 0-10 \
    --providers aws_earth \
    --max-results 5 \
    --format ids
```

```python
# CQL2 filter
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --start-date 2024-01-01 --end-date 2024-06-01 \
    --providers planetary_computer \
    --cql2 "platform = 'sentinel-2b'" \
    --max-results 5 \
    --format table
```

```python
# No-cache — force fresh results
!PyGeoFetch search run \
    --bbox "-74.1,40.6,-73.7,40.9" \
    --cloud-cover 0-10 \
    --providers aws_earth \
    --max-results 3 \
    --no-cache \
    --format table
```

## 7. Pipeline Commands

```python
# Write a test pipeline
pipeline_yaml = """\
name: cli-demo-pipeline
description: Demo pipeline for CLI notebook

steps:
  - search:
      providers: [aws_earth]
      bbox: "-74.1,40.6,-73.7,40.9"
      date_range: last_30_days
      cloud_cover: "0-20"
      max_results: 5

  - download:
      parallel: 1
      output: ./output/pipeline_output/
      verify_checksum: false
"""

from pathlib import Path
pl_path = Path("output/cli_pipeline.yaml")
pl_path.write_text(pipeline_yaml)
print(f"Pipeline written → {pl_path}")
```

```python
# Validate without running
!PyGeoFetch pipeline validate output/cli_pipeline.yaml
```

```python
# Run the pipeline
!PyGeoFetch pipeline run output/cli_pipeline.yaml
```

```python
# Schedule it
!PyGeoFetch pipeline schedule output/cli_pipeline.yaml --name "cli-demo"
```

```python
# List all scheduled
!PyGeoFetch pipeline list-scheduled
```

```python
# Execution history
!PyGeoFetch pipeline history --limit 5
```

```python
# Unschedule
!PyGeoFetch pipeline unschedule cli-demo
```

## Complete CLI Cheat Sheet

```python
cheatsheet = """
═══════════════════════════════════════════════════════════════
  PyGeoFetch CLI Cheat Sheet
═══════════════════════════════════════════════════════════════

SYSTEM
  PyGeoFetch --version
  PyGeoFetch doctor
  PyGeoFetch status [--json]

AUTH
  PyGeoFetch auth add PROVIDER [--username U] [--password P] [--api-key K]
  PyGeoFetch auth login PROVIDER          # interactive
  PyGeoFetch auth list [--json]
  PyGeoFetch auth test PROVIDER
  PyGeoFetch auth remove PROVIDER [--yes]
  PyGeoFetch auth export [--output FILE]

PROVIDERS
  PyGeoFetch providers list [--auth|--no-auth] [--capabilities sar] [--json]
  PyGeoFetch providers info PROVIDER [--json]
  PyGeoFetch providers search "TERM"

SEARCH
  PyGeoFetch search run \\
    --bbox "minlon,minlat,maxlon,maxlat"      # or --geometry-file file.geojson
    --start-date YYYY-MM-DD
    --end-date   YYYY-MM-DD
    --cloud-cover 0-20
    --providers aws_earth,copernicus
    --satellites Sentinel-2
    --sort-by cloud_cover --sort-order asc
    --max-results 50
    --cql2 "eo:cloud_cover < 5"
    --format table|json|stac|geojson|geoparquet|csv|ids
    --output results.geojson
    --no-cache
    --timeout 60
    --on-provider-failure skip

DOWNLOAD
  PyGeoFetch download run \\
    --from-search results.geojson            # required
    --output ./data/                         # output dir
    --parallel 4
    --retry 5 --retry-delay 2.0
    --verify-checksum
    --resume
    --bandwidth-limit 10MB
    --priority high
    --bands "B02,B03,B04"
    --post-process "reproject:EPSG:4326,compress:lzw,cog"
    --on-failure skip
    --max-items 10
    --overwrite
    --notify webhook:https://hooks.slack.com/...
    --json

CACHE
  PyGeoFetch cache stats [--json]
  PyGeoFetch cache clear [--provider P] [--older-than 7d] [--dry-run]
  PyGeoFetch cache ttl show|set SECONDS
  PyGeoFetch cache location
  PyGeoFetch cache prune --max-size 1GB

PIPELINE
  PyGeoFetch pipeline run FILE [--step STEP_NAME]
  PyGeoFetch pipeline validate FILE
  PyGeoFetch pipeline schedule FILE --name NAME [--cron "0 6 * * 1"]
  PyGeoFetch pipeline list-scheduled [--json]
  PyGeoFetch pipeline unschedule NAME
  PyGeoFetch pipeline history [--limit 20]
  PyGeoFetch pipeline logs NAME [--tail 50]

CONFIG
  PyGeoFetch config show [--json]
  PyGeoFetch config get KEY
  PyGeoFetch config set KEY VALUE
  PyGeoFetch config path
  PyGeoFetch config reset

COMPLETION
  PyGeoFetch --install-completion bash|zsh|fish
═══════════════════════════════════════════════════════════════
"""
print(cheatsheet)

# Save it
Path("output/pygeofetch_cheatsheet.txt").write_text(cheatsheet)
print("Cheatsheet saved → output/pygeofetch_cheatsheet.txt")
```
