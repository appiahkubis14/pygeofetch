## 7. Spectral Index Commands — All 17

```python
print("PyGeoFetch index commands (17 indices):")
print()
index_cmds = [
    ("ndvi",      "PyGeoFetch index ndvi  --red B04.tif --nir B08.tif"),
    ("evi",       "PyGeoFetch index evi   --blue B02.tif --red B04.tif --nir B08.tif"),
    ("savi",      "PyGeoFetch index savi  --red B04.tif --nir B08.tif --soil-l 0.5"),
    ("ndwi",      "PyGeoFetch index ndwi  --green B03.tif --nir B08.tif"),
    ("mndwi",     "PyGeoFetch index mndwi --green B03.tif --swir1 B11.tif"),
    ("ndbi",      "PyGeoFetch index ndbi  --nir B08.tif --swir1 B11.tif"),
    ("ndsi",      "PyGeoFetch index ndsi  --green B03.tif --swir1 B11.tif"),
    ("ndmi",      "PyGeoFetch index ndmi  --nir B08.tif --swir1 B11.tif"),
    ("nbr",       "PyGeoFetch index nbr   --nir B08.tif --swir2 B12.tif"),
    ("dnbr",      "PyGeoFetch index dnbr  --pre-nir B08.tif --pre-swir2 B12.tif --post-nir B08p.tif --post-swir2 B12p.tif"),
    ("tct",       "PyGeoFetch index tct   --blue B02.tif --green B03.tif --red B04.tif --nir B08.tif --swir1 B11.tif --swir2 B12.tif"),
    ("pca",       "PyGeoFetch index pca   B02.tif B03.tif B04.tif B08.tif --components 3"),
    ("texture",   "PyGeoFetch index texture B08.tif --window 7 --features contrast,homogeneity,energy"),
    ("lst",       "PyGeoFetch index lst   B10.tif --emissivity 0.97 --sensor landsat8"),
    ("albedo",    "PyGeoFetch index albedo B02.tif B03.tif B04.tif B08.tif B11.tif B12.tif"),
    ("band-math", "PyGeoFetch index band-math B04.tif B08.tif --expr '(B[1]-B[0])/(B[1]+B[0]+1e-6)'"),
    ("stack",     "PyGeoFetch index stack B02.tif B03.tif B04.tif"),
]
for name, cmd in index_cmds:
    print(f"  [{name}]")
    print(f"    {cmd}")
    print()
```

## 8. Post-Processing Commands

```python
print("PyGeoFetch post commands:")
print()
post_cmds = [
    "PyGeoFetch post vectorize       ndvi.tif --threshold 0.3 --format geojson",
    "PyGeoFetch post smooth          polygons.geojson --tolerance 0.5",
    "PyGeoFetch post regularize      buildings.geojson",
    "PyGeoFetch post zonal-stats     ndvi.tif parcels.geojson --output stats.csv",
    "PyGeoFetch post buffer          roads.geojson --distance 15",
    "PyGeoFetch post centroids       polygons.geojson",
    "PyGeoFetch post geometry-metrics polygons.geojson",
    "PyGeoFetch post compress        scene.tif --method lzw",
    "PyGeoFetch post cog             scene.tif --compress deflate --blocksize 512",
]
for cmd in post_cmds:
    print(f"  {cmd}")
```

## 9. SAR Processing Commands

```python
print("PyGeoFetch sar commands:")
print()
sar_cmds = [
    "PyGeoFetch sar despeckle  sar.tif --filter lee            --window 5",
    "PyGeoFetch sar despeckle  sar.tif --filter enhanced_lee   --window 7",
    "PyGeoFetch sar despeckle  sar.tif --filter frost",
    "PyGeoFetch sar despeckle  sar.tif --filter gamma",
    "PyGeoFetch sar despeckle  sar.tif --filter boxcar",
    "",
    "PyGeoFetch sar calibrate  sar.tif --output-type sigma0 --db",
    "PyGeoFetch sar calibrate  sar.tif --output-type gamma0 --db",
    "PyGeoFetch sar calibrate  sar.tif --output-type beta0  --linear",
    "",
    "PyGeoFetch sar flood-map  post.tif --threshold -15.0",
    "PyGeoFetch sar flood-map  post.tif --reference pre.tif",
    "",
    "PyGeoFetch sar coherence  slc1.tif slc2.tif --window 7",
]
for cmd in sar_cmds:
    if cmd:
        print(f"  {cmd}")
    else:
        print()
```

## 10. Processing Pipeline CLI

```python
print("Processing pipeline commands:")
print()
proc_cmds = [
    "# Generate starter YAML template",
    "PyGeoFetch proc-pipeline template ndvi",
    "PyGeoFetch proc-pipeline template change_detection",
    "PyGeoFetch proc-pipeline template flood_map",
    "PyGeoFetch proc-pipeline template urban_mapping",
    "PyGeoFetch proc-pipeline template sar_analysis",
    "PyGeoFetch proc-pipeline template land_cover",
    "",
    "# Validate without running",
    "PyGeoFetch proc-pipeline validate ndvi_pipeline.yaml",
    "",
    "# Run processing pipeline",
    "PyGeoFetch proc-pipeline run ndvi_pipeline.yaml --input scene.tif --output-dir ./out/",
    "",
    "# Data pipeline (search + download)",
    "PyGeoFetch pipeline run weekly.yaml",
    "PyGeoFetch pipeline validate weekly.yaml",
    "PyGeoFetch pipeline schedule weekly.yaml --name ndvi-weekly",
    "PyGeoFetch pipeline list-scheduled",
    "PyGeoFetch pipeline history --limit 10",
    "PyGeoFetch pipeline unschedule ndvi-weekly",
]
for cmd in proc_cmds:
    if cmd.startswith("#"):
        print(f"
  {cmd}")
    elif cmd:
        print(f"    {cmd}")
```
